```python
import tkinter as tk
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
```

```python
bg1 = tk.Tk()
ask_theta = tk.Label(bg1, text="How many data points for polar(theta) angle?")
ask_phi = tk.Label(bg1, text="How many data points for azimuthal(phi) angle?")
ask_phi.grid(row=0, column=0)
ask_theta.grid(row=1, column=0)

theta_num = tk.IntVar()
phi_num = tk.IntVar()
get_theta_num = tk.Entry(bg1, textvariable=theta_num)
get_phi_num = tk.Entry(bg1, textvariable=phi_num)
get_theta_num.grid(row=0, column=1)
get_phi_num.grid(row=1, column=1)
def get_theta(self):
    theta_num = get_theta_num.get()
    print(theta_num)
def get_phi(self):
    phi_num = get_phi_num.get()
    print(phi_num)
get_theta_num.bind('<Return>', get_theta)
get_phi_num.bind('<Return>', get_phi)

tk.mainloop()
```

```python
# convert to python data str
theta_num = theta_num.get()
phi_num = phi_num.get()
```

```python
# now creating two arrays for the sphere construction
theta_array, phi_array = np.mgrid[0.0:np.pi:theta_num*1j, 0.0:2.0*np.pi:phi_num*1j]
x_array = np.sin(theta_array) * np.cos(phi_array)
y_array = np.sin(theta_array) * np.sin(phi_array)
z_array = np.cos(theta_array)
meridian_y = np.sin(np.linspace(0, np.pi, theta_num))
meridian_z = np.cos(np.linspace(0, np.pi, theta_num))
meridian_x = [0]*len(meridian_x)
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot3D(meridian_x, meridian_y, meridian_z) # a reference line for rotation
ax.plot_surface(x_array, y_array, z_array, alpha=0.2)
ax.scatter(x_array, y_array, z_array, s=10)

ax.set_xlim([-1,1])
ax.set_ylim([-1,1])
ax.set_zlim([-1,1])
# plt.tight_layout()
ax.view_init(25, 0)
plt.show()
```
